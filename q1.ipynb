```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def bernoulli_pdf (theta_head, num_heads, num_tails):
    
    theta_tail = 1 - theta_head
    
    return (theta_head ** num_heads) * (theta_tail ** num_tails)
```

```python
theta = np.arange(0.0, 1.0, 0.01)
```

```python
bern_pdf_a = np.array([bernoulli_pdf(theta_i,6,4) for theta_i in theta])
```

```python
plt.title('Likelihood plot for question 1_a / 6 Heads, 4 Tails, N=10')
plt.xlabel('Q')
plt.ylabel('Likelihood')
plt.plot(theta,bern_pdf_a)
plt.axvline(x = theta[np.argmax(bern_pdf_a)], color = 'red')
```
