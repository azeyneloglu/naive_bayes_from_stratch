```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import special
```

```python
theta = np.arange(0.0, 1.0, 0.01)
```

```python
def bernoulli_pdf (theta_head, num_heads, num_tails):
    
    theta_tail = 1 - theta_head
    
    return (theta_head ** num_heads) * (theta_tail ** num_tails)
```

```python
def beta_pdf (theta_head, beta_heads, beta_tails, data_heads, data_tails):
    
    a = beta_heads + data_heads
    b = beta_tails + data_tails
    
    return (1 / special.beta(a,b)) * bernoulli_pdf(theta_head,a-1,b-1)
```

```python
betas = np.array([beta_pdf(theta_i,3,3,6,4) for theta_i in theta])
```

```python
plt.title('MAP Estimate for Question 1_e / 6 Heads, 4 Tails with prior beta(3,3)')
plt.xlabel('Q')
plt.ylabel('Posterior')
plt.plot(theta,betas)
plt.axvline(x = theta[np.argmax(betas)], color = 'red')
```

```python
#map estimate is a little different than mle estimate because we took our prior beliefs of theta into account
#for the map estimation.
theta[np.argmax(betas)]
```

```python
betas_infinite = np.array([beta_pdf(theta_i,3,3,600,400) for theta_i in theta])
```

```python
plt.title('MAP Estimate for Question 1_e / 600 Heads, 400 Tails with prior beta(3,3)')
plt.xlabel('Q')
plt.ylabel('Posterior')
plt.plot(theta,betas_infinite)
plt.axvline(x = theta[np.argmax(betas_infinite)], color = 'red')
```

```python
# As we increase the data size effect of prior becomes negligible and MAP estimate has the same value as the MLE estimate.
```
